```python
import os
import cv2
import tqdm

video_path = '../crabs_dataset/eval_videos/fullvideo_test6/sample6.mp4'
video_name = 'sample6'
save_dir = '../crabs_dataset/eval_videos/fullvideo_test6/parse_nearframe'
os.makedirs(save_dir, exist_ok=True)

vs = cv2.VideoCapture(video_path)
frame_count = int(vs.get(cv2.CAP_PROP_FRAME_COUNT))
fps = int(vs.get(cv2.CAP_PROP_FPS))

inter_frame = 3
outer_frame = 30*30

for idx in tqdm.tqdm(range(frame_count)):
    ret, frame = vs.read()
    if idx%outer_frame>=0 and idx%outer_frame<=6*inter_frame:
        foldname = '{:08d}'.format((idx//outer_frame)*outer_frame)
        foldpath = os.path.join(save_dir, foldname)
        os.makedirs(foldpath, exist_ok=True)
        if idx%inter_frame==0:
            savepath = os.path.join(foldpath, video_name + '_' + '{:08d}'.format(idx)+'.png')
            cv2.imwrite(savepath, frame)
```

```python
import os
import shutil
import pandas as pd

basepath = '../crabs_dataset/fashion-dataset/fashion-dataset/'
styles_csv_path = os.path.join(basepath, 'styles.csv')

styles_csv = pd.read_csv(styles_csv_path)
styles_csv
```

```python
styles_csv.loc[0]['baseColour']
```

```python
masterCategory = []
subCategory = []
articleType = []
baseColour = []
for i in range(len(styles_csv)):
    data = styles_csv.loc[i]
    if data['masterCategory'] not in masterCategory:
        masterCategory.append(data['masterCategory'])
    if data['subCategory'] not in subCategory:
        subCategory.append(data['subCategory'])
    if data['articleType'] not in articleType:
        articleType.append(data['articleType'])
    if data['baseColour'] not in baseColour:
        baseColour.append(data['baseColour'])

print(masterCategory)
print(subCategory)
print(articleType)
print(baseColour)
```

```python
import os
import cv2
import tqdm

basepath = '../crabs_dataset/livecommerce/ssg_videos/'
# video_name = 'sample6'
namelist = sorted(os.listdir(basepath))
save_dir = '../crabs_dataset/livecommerce/ssg_videos_parse_nearframe/'
inter_frame = 5
outer_frame = 30*30

os.makedirs(save_dir, exist_ok=True)

for video_name in namelist:
    video_path = os.path.join(basepath, video_name)
    vs = cv2.VideoCapture(video_path)
    frame_count = int(vs.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = int(vs.get(cv2.CAP_PROP_FPS))
    video_dir = os.path.join(save_dir, video_name[:-4])

    for idx in tqdm.tqdm(range(frame_count)):
        ret, frame = vs.read()
        if idx > 3*outer_frame and idx%outer_frame>=0 and idx%outer_frame<=6*inter_frame:
            
            foldname = video_name[:-4] + '_' +'{:08d}'.format((idx//outer_frame)*outer_frame)
            foldpath = os.path.join(video_dir, foldname)
            os.makedirs(foldpath, exist_ok=True)
            if idx%inter_frame==0:
                savepath = os.path.join(foldpath, video_name[:-4] + '_' + '{:08d}'.format(idx)+'.png')
                cv2.imwrite(savepath, frame)
```
